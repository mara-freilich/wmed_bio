### Mathematical modeling approach to phytoplankton biogeography

$\frac{dP}{dt} = \underbrace{G_P}_\text{maximum growth rate} \underbrace{(1-e^{-\gamma I(z)/I_0})}_\text{limitation by light}\underbrace{\frac{N}{k_P+N}}_\text{limitation by nutrients}P-\underbrace{m_1 P - m_2 P^2}_\text{mortality}$

### Monod function
$\frac{N}{k+N}$
- Derived from chemical kinetics
- Parameter is a half saturation constant
    - This parameter determines growth rate at low nutrient concentration
    - Small cells have lower half saturation constant due to surface area to volume ratio

If N = k

$\frac{N}{k+N} = \frac{k}{2k} = \frac{1}{2}$

```python
N = np.linspace(0,50,500)
kP = 3 # umol
kS = 0.5 # umol

GP = 1 # per day
GS = 1 # per day

fig,ax = plt.subplots(1,1,figsize = (6,5))
ax.plot(N,GP*N/(kP+N),label = 'P')
ax.plot(N,GS*N/(kS+N),label = 'S')
ax.legend()
ax.set_xlabel('nutrient')
ax.set_ylabel('growth rate (day$^{-1}$)')
```

```python
N = np.linspace(0,50,500)
kP = 3 # umol
kS = 0.5 # umol
GP = 1 # per day
GS = 1 # per day

fig,ax = plt.subplots(1,1,figsize = (6,5))
ax.plot(N,GP*N/(kP+N),label = 'P')
ax.plot(N,GS*N/(kS+N),label = 'S')
ax.plot([0,50],[0.5,0.5],color = 'grey')
ax.plot([kP,kP],[0,1],color = 'tab:blue',ls = '--'); ax.plot([kS,kS],[0,1],color = 'tab:orange',ls = '--')
ax.legend()
ax.set_xlabel('nutrient')
ax.set_ylabel('growth rate (day$^{-1}$)')
```

### Maximum growth rate

- Maximum growth rate depends on cell size
- as phytoplankton species become larger, their ability to take up nitrogen, when available in large concentrations, increases

### Light limitation

The typical euphotic depth (or depth where there is enough light to grow) is 100 meters, but it varies globally. The euphotic depth is usually defined as 1% of the surface light level.

<img src="talley_dpo.jpg" width=400 height=400 />

```python
z = np.linspace(-300,0)
kpar = 0.04 # per meter
I0 = 1000 # watts per square meter

fig,ax = plt.subplots(1,1,figsize = (6,5))
ax.plot(I0*np.exp(kpar*z),z,color = 'k')
ax.set_ylabel('z (meters)')
ax.set_xlabel(r'irradiance in W/m$^2$')
```

```python
z = np.linspace(-300,0)
kpar = 0.04 # per meter
I0 = 1000 # watts per square meter
I = I0*np.exp(kpar*z)

fig,ax = plt.subplots(1,1,figsize = (6,5))
ax.plot(I/I0,z,color = 'grey',label = '$I/I_0$')
ax.set_ylabel('z (meters)')

gamma = 1 # unitless
ax.plot(1-np.exp(-gamma*I/I0),z,label = r'$\gamma$ = 1')
gamma = 10 # unitless
ax.plot(1-np.exp(-gamma*I/I0),z,label = r'$\gamma$ = 10')
ax.legend()
```

### Mortality

There are two terms used for mortality. 
- The linear term $m_1P$ represents respiration.
- The non-linear term $m_1P^2$ represents density-dependent loss, for example due to predation
